### Problem 1

```python
students = [
    {"name":"Anand","marks":85},
    {"name":"Rahul","marks":92},
    {"name":"Amit","marks":76},
    {"name":"Neha","marks":89}
]

#1️⃣ Sort students by marks using lambda

students.sort(key=lambda x : x["marks"])
print(students)
```

- 2️⃣ Find top scorer

```python
max_mark=0
top_scorer=""
for i in range(len(students)):
    if students[i]['marks'] > max_mark:
        max_mark = students[i]['marks']
        top_scorer = students[i]['name']
print("Top scorer is {} with marks {}".format(top_scorer,max_mark))
```

```python
top_student = max(students, key=lambda x: x['marks'])

print(top_student)
```

- 3️⃣ Calculate average marks

```python
sum=0
for i in range(len(students)):
    sum += students[i]['marks']

average = sum/len(students)
print(average)
```

- 4️⃣ Return list of students scoring above average

```python
studs = filter(lambda x : x['marks']>average , students)
print(list(studs))
```

### Problem 2

```python
sentence = "python is great and python is easy and python is powerful"
```

- 1️⃣ Count frequency of each word

```python
words = sentence.split()

freq = {}

for word in words:
    freq[word] = freq.get(word,0)+1

print(freq)
```

- Sort dictionary by frequency using lambda

```python
lst = freq.items()

sorted_lst = sorted(lst,key=lambda x : x[1])

print(sorted_lst)
```

-  Print the most common word

```python
common = max(lst,key=lambda x : x[1])

print(common)

print(max(freq.values()))
```

```python
for i in range(len(sorted_lst)):
    if(sorted_lst[i][1] == max(freq.values())):
        print(sorted_lst[i][0])
```

### Problem 6

```python
employees = [
    {"name":"Anand","salary":50000,"experience":3},
    {"name":"Rahul","salary":70000,"experience":5},
    {"name":"Amit","salary":45000,"experience":2},
    {"name":"Neha","salary":80000,"experience":6}
]
```

1️⃣ Increase salary 10% if experience > 3 years

```python
for key in range(len(employees)):
    if(employees[key]['experience'] > 3):
        employees[key]['salary'] *= 1.10

print(employees)
```

- 2️⃣ Sort employees by salary using lambda
  

```python
st = sorted(employees,key=lambda x : x['salary'])

print(st)
```

- 3️⃣ Find employee with highest salary

```python
max_sal = max(employees,key=lambda x : x['salary'])

print(max_sal)
```

- 4️⃣ Create list of employees earning more than 60000

```python
lst = list(filter(lambda x : x['salary']>60000,employees))

print(lst)
```

```python
spend = {}

for i in range(len(orders)):
    username = orders[i]['user']
    userprice = orders[i]['price']

    if username in spend:
        spend[username] += userprice
    else :
        spend[username] = userprice

print(spend)
```

```python
lst = spend.items()

sorted_lst = sorted(lst,key=lambda x : x[1])

dic = dict(sorted_lst)
print(dic)
```

```python
max_ele = max(sorted_lst,key=lambda x : x[1])

print(max_ele)
```
