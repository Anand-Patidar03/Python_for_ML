## Generator

```python
def square(n):
    for i in range(n):
        yield i**2

type(square(6))
```

```python
for i in square(5):
    print(i)
```

#### Difference
- to create a iterator we use iter() and to generator we use function along with yeild keyword as above
- generator saves the local variable
- generator help to write fast and compact model
