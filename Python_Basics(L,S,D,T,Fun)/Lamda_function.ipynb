# Lamda Function

- A lambda function is a small anonymous function (a function without a name).

It is used when:

the function is very short

used once

passed as an argument to another function

Syntax:

lambda arguments : expression

Important rules:

Only one expression

No return keyword

The result of expression is returned automatically

Equivalent to a one-line function.

```python
square = lambda x : x*x

print(square(5))
```

```python
add = lambda a,b : a+b

print(add(3,4))
```

```python
print((lambda x : x**2)(4))
```

```python
nums = [1,2,3,4,5,6]

sqr = map(lambda x : x**2 , nums)

print(list(sqr))
```

```python
nums = [1,2,3,4,5,6,7,8,9]

evens = filter(lambda x : x%2==0 , nums)

print(list(evens))
```

```python
players = [("Virat",18),("Rohit",45),("MSD",7),("Sachin",10)]

players.sort(key = lambda x : x[1])

print(players)
```

```python
string_len = lambda s : len(s)

string_len("Anand Patidar")
```

```python
print(string_len("Anand Patidar"))
```

```python
nums = [1,2,3,4,5]
lst = map(lambda x : x*x , nums)

print(list(lst))
```

```python
lst = ["anand","patidar","is","learning","python"]

lst.sort(key=lambda x : len(x))

print(lst)
```

```python
players = [{"age":18},{"age":45},{"age":7},{"age":10}]

players.sort(key = lambda x : x['age'])

# print(sorted(players,key=lambda x:x['age']))          #alternate method using sorted

print(players)
```

```python
# find the maximum value in dictionary

data = {'a':10,"b":5,"c":15,"d":20}   

max(data,key=lambda x : data[x])
```

```python
# grouping a list of string based on their first letter

from itertools import groupby

words = ["apple","banana","cherry","mango","berry","papaya"]

grps = groupby(sorted(words,key=lambda x : x[0]))

for key,group in  grps:
    print(key,list(group))
```

### list comprehension

- it is a way to concise the list in single line.
- It consist of brackets containing a expression followed by a for clause  and then zero or more for or if clauses

```python
nums = [1,2,3,4,5,6,7,8,9]

evens = [x**2 for x in nums if x%2==0]

print(evens)
```
