#### Matrix Multiplication

```python
lst1 = [-2,-1, 3]
lst2 = [-6 ,-3 ,9]

arr = np.array([lst1,lst2])

arrT = np.transpose(arr)

print(arr)
print(arrT)

mul = np.matmul(arr,arrT)          #Matrix Multiplication
mul1 = np.dot(arr,arrT)            #Matrix Multiplication

print(mul1)
print(mul)
```

## Linear Algebra

```python
lst1 = [4,7]
lst2 = [2,7]

arr = np.array([lst1,lst2])

print(np.linalg.det(arr))
print("\n")
print(np.linalg.inv(arr))
print("\n")
print(np.linalg.eig(arr))
```
