# Dataset 1

```python
import numpy as np

np.random.seed(42)

n = 120  # number of rows

order_id = np.arange(1000, 1000 + n)

customer_id = np.random.randint(100, 200, size=n)

product_category = np.random.choice([0, 1, 2, 3], size=n, p=[0.3, 0.3, 0.2, 0.2])

price = np.round(np.random.uniform(200, 5000, size=n), 2)

quantity = np.random.randint(1, 6, size=n)

discount = np.round(np.random.choice([0, 0.1, 0.2, 0.3, 0.5],size=n,p=[0.4, 0.2, 0.2, 0.1, 0.1]), 2)

rating = np.random.randint(1, 6, size=n)

delivery_days = np.random.randint(1, 10, size=n)

is_returned = (rating <= 2).astype(int)

# Final dataset
data = np.column_stack((
    order_id,
    customer_id,
    product_category,
    price,
    quantity,
    discount,
    rating,
    delivery_days,
    is_returned
))

print(data.shape)  # (120, 9)
```

```python
print(data)
```

🔹 Medium Level (Warm-up)
Q1. High Value Orders

Find all orders where total purchase value (price × quantity) is greater than ₹10,000.

Hint:
Create a new array for total value → use boolean masking.

Q2. Electronics Performance

For Electronics category (0), compute:

Average rating
Average delivery days

Hint:
Filter rows first → then apply aggregation functions.

```python
# Q1
total_value = data[data[:, 3] * data[:, 4] > 10000]
print(total_value)

# Q2
elec = data[data[:, 2] == 0]

avg_rating = elec[:, 6].mean()
print(avg_rating)

avg_days = elec[:, 7].mean()
print(avg_days)
```

```python
disc = data[data[:,5] >= 0.2]
print(disc)
```

```python
nodisc = data[data[:,5] == 0]
print(nodisc)
```

```python
ttl_valdisc = (disc[:,3] * disc[:,4]).mean()
print(ttl_valdisc)

ttl_valnodisc = (nodisc[:,3] * nodisc[:,4]).mean()
print(ttl_valnodisc)
```

```python
cust , count = np.unique(data[:,1] , return_counts=True)      # Return array and count number

idx = np.argsort(count)

print(idx)

top_five = idx[-5:]
print(top_five)
```

```python
top_customers = cust[top_five]
top_counts = count[top_five]

print(top_customers)
print(top_counts)
```

```python
arr = data[:,2]

cat = np.unique(arr)
res = []

for c in cat:
    ele = data[data[:,2] == c]
    ttl_rev = (ele[:,3] * ele[:,4])
    summ = ttl_rev.sum()
    res.append(summ)

type(res)
ans = np.array(res)
print(ans)
```
