- 👉 Task:

- Identify students whose average score is less than 75

```python
avgg = np.average(data,axis = 1)
print(avgg)
a = data[avgg > 75]
print(a)
```

- Normalize the dataset between 0 and 1 using: Formula (x-xmin)/(xmax-xmin)

```python
normalized_dataset = (data - np.min(data)) / (np.max(data) - np.min(data))

print(normalized_dataset)
```

- 👉 Task:
- Create a boolean array where marks ≥ 75 are True.

```python
bools = data>75
upArr = np.where(~bools , 70 , data)

print(bools)
print(upArr)
```

#### New dataset 

```python
data = np.array([
    [85, np.nan, 92],
    [88, 76, np.nan],
    [np.nan, 85, 91]
])
```

- 👉 Task:

Replace NaN with column-wise mean

```python
col_mean = np.nanmean(data,axis = 0)
idx = np.where(np.isnan(data))
data[idx] = np.take(col_mean,idx[1])

print(idx)

print(data)
```

- Standardize each feature using:

```python
std_data = (data - np.mean(data)) / np.std(data)

print(std_data)
```

👉 Task:

- Find values greater than mean + 2*std
- Replace them with mean

```python
cond = np.mean(data) +  np.std(data)
print(cond)
new_data = np.where(data > cond,np.mean(data),data)
print(new_data)
```

```python
data = np.array([
    [85, 78, 92],
    [88, 76, 95],
    [90, 85, 91],
    [70, 60, 65]
])
```

```python
import numpy as np
row_mean = np.mean(data, axis=1, keepdims=True)
row_std = np.std(data, axis=1, keepdims=True)
print(row_mean)
print(row_std)
normalized_data = (data - row_mean) / row_std

print(normalized_data)
```
