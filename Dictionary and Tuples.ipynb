```python
sqr = {x : x*x for x in range(6)}
print(sqr)
```

```python
dic = {1:1,2:4,3:9,4:16}
print(dic)
```

```python
string = "banana"
freq = {}

for ch in string:
    if ch in freq:
        freq[ch] += 1

print(freq)
```

### Nested Discrionary

```python
stud1 = {"name":"Anand","age":21}
stud2 = {"name":"Vijay","age":22}
stud3 = {"name":"Ajay","age":23}
print(stud1)
print(stud2)
print(stud3)
```

```python
friend = {"me":stud1,"fr1":stud2,"fr2":stud3}
print(friend)
```

```python
print(stud1["name"])
```

```python
print(friend["me"])
```

```python
print(friend["me"]["age"])
```

```python
for i in friend.values():
    print(i["name"] , ":", i["age"])
```

# Tuples

- A Tuple is an ordered collection of elements that cannot be changed.

Key properties:

Ordered

Immutable (cannot modify)

Allows duplicates

Can store mixed data types

```python
tup = tuple()
type(tup)
```
