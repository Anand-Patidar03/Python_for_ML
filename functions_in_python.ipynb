# Functions

-A function is a reusable block of code that performs a specific task.

Instead of repeating code, we define once and reuse many times.

```python
def greet():
    print("Hello Anand")

greet()
```

```python
def namaste(name):
    print("Namaste",name)
namaste("Anand Ji")
namaste("Patidar Ji")
```

```python
def square(a):
    print("Squar of",a,"is",a*a)
square(5)
```

```python
def add(a,b=8):
    return a+b

print(add(b=0,a=5))
```

```python
def mul(a,b):
    return a*b

mul(5,2)
```

```python
def greet(name = "Guest"):
    print("Hello",name)
greet()
greet("Anand")
```

```python
def power(a,b=2):
    print(a,"power",b,"is",a**b)
power(5)
```

```python
def arithmetic(a,b):
    return a+b,a-b,a*b,a/b
arithmetic(4,2)    # return in form of tuple

summ,diff,mul,div = arithmetic(4,2)
print("Sum is",summ)
print("difference is",diff)
print("product is",mul)
print("Division is",div)
```

```python
def min_max(arr):
    return min(arr),max(arr)

arr = [5,3,7,2,4]
minm,maxm = min_max(arr)
print("Minimum number is",minm)
print("Maximum number is",maxm)
```

```python
def average(*numbers):
    sum=0
    for i in numbers:
        sum = sum + i
    avg = sum/len(numbers)
    return avg

numbers = [1,2,3,4,5,6]
avg1 = average(*numbers)
avg2 = average(5,6,7,8)
print("Average of numbers is",avg1)
print("Average of numbers is",avg2)
```

```python
def student(**data):
    print(data)

student(Name="Anand",Age=21)
type(student(Name="Anand",Age=21))   #NoneType
```

```python
def info(**data):
    for k,v in data.items():
        print(k,v)

info(name="Anand", branch="CSE", age=21)
```

```python
def fact(n):
    f = 1
    for i in range(1,n+1):
        f *= i
    return f

factorial = fact(5)
print("Factorial of 5 is ",factorial)
```

```python
def cnt_vowels(string):
    cnt=0
    for i in range(len(string)):
        if(string[i] in "aeiou"):
            cnt += 1

    return cnt

print("Total vowels are ",cnt_vowels("anand patidar"))
```
