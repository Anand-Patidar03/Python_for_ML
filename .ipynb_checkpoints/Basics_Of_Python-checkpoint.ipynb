### Conditional Statements


```python
x = 10

if x>5:
    print("greater than 5")
else:
    print("Less than 5")
```

```python
x = 10

if(x>5):
    print("greater than 5")
else:
    print("Less than 5")
```

```python
x = int(input("Enter a number to check even or odd : "))

if(x%2 == 0):
    print("x is Even")
else  :
    print("x is Odd")
```

- while loop

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

### String and Boolean

```python
name = "Anand123 "   #if space comes neither isalnum nor isalpha
print(name.isalnum())
print(name.isalpha())
```

```python
name = "Anand Patidar"
print(name.isdigit())
print(name.istitle())
print(name.islower())
print(name.isupper())
print(name.isnumeric())
print(name.isidentifier())
print(name.isspace())
print(name.isascii())
```

```python
print(name.startswith('a'))
print(name.endswith("r"))
```
