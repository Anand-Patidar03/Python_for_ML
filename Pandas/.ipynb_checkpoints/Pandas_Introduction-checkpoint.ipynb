## Pandas DataFrame Manipulation

```python
import pandas as pd

data = {
    'Name': ['Aman', 'Riya', 'Raj', 'Simran', 'Karan'],
    'Age': [20, 21, 19, 22, 20],
    'Marks': [85, 90, 78, 92, 88],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Pune', 'Mumbai']
}

df = pd.DataFrame(data)
```

```python
print(df)
```

```python
df.head()
```

```python
rws = df.iloc[0:3,:]
print(rws)
```

```python
df.head(3)
```

```python
df[['Name','Marks']]
```

```python
df[df['Marks'] > 85]
```

```python
df['Passed'] = np.where(df['Marks'] >= 80 , 'Pass' , 'Fail')
```

```python
df.head()
```

```python
avgg = np.mean(df['Marks'])
print(avgg)
```

## Another Dataset

```python
data = {
    'Product': ['Laptop', 'Phone', 'Tablet', 'Laptop', 'Phone'],
    'Price': [70000, 20000, 30000, 75000, 22000],
    'Quantity': [3, 5, 2, 4, 6],
    'City': ['Delhi', 'Delhi', 'Mumbai', 'Mumbai', 'Delhi']
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df['Total Sales'] = df['Price'] * df['Quantity']

df.head()
```

```python
total_sales = df.groupby('Product')['Total Sales'].sum()
print(total_sales)
```

```python
df.head()
```

```python
prods = df[df['City'] == 'Delhi']

prods.head()
```

```python
srt = df.sort_values('Price',ascending = False)
print(srt)
```

```python
hgt_sale = df.max()
print(hgt_sale)
```

```python
print(df.loc[df['Total Sales'].idxmax()])
```

## Another Dataset

```python
data = {
    'Name': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Department': ['IT', 'HR', 'IT', 'HR', 'IT', 'HR'],
    'Salary': [60000, 50000, 70000, 52000, 65000, 58000],
    'Experience': [2, 3, 5, 1, 4, 2]
}

df = pd.DataFrame(data)
```

```python
df.head(6)
```

- Find average salary per department

```python
avg_sal = df.groupby('Department')['Salary'].mean()

print(avg_sal)
```

- Find employee with highest salary in each department

```python
idx = df.groupby('Department')['Salary'].idxmax()
emp = df.loc[idx]
print(emp)
```

- Add a column "Senior"

- Condition:

- Experience >= 3 → Yes
- Else → No

```python
df['Senior'] = np.where(df['Experience'] >= 3 , "Yes", "No")

df.head()
```

- Filter employees:
- Salary > 55000 AND Department = IT

```python
emp = df[((df['Salary'] > 57000) & (df['Department'] == 'IT'))]
print(emp)
```

- Rank employees based on Salary

```python
df.head(6)
```

```python
df['Rank'] = df['Salary'].rank(ascending = False)
```

```python
df.head()
```

## Another Dataset

```python
data = {
    'OrderID': [101,102,103,104,105,106],
    'Customer': ['Aman','Riya','Aman','Raj','Riya','Simran'],
    'Amount': [500,700,300,400,900,1000],
    'Status': ['Completed','Pending','Completed','Completed','Pending','Completed']
}

df = pd.DataFrame(data)
```

```python
df.head(7)
```

- Total amount spent by each customer

```python
amt = df.groupby('Customer')['Amount'].sum()
print(amt)
```

- Count number of completed orders

```python
print((df['Status'] == 'Completed').sum())
```

- Find customers who spent more than 800 total

```python
amt.head()
```

```python
mt = df.groupby('Customer')['Amount'].sum()
mt[mt>800]
```

- Find average order amount

```python
avgg = df['Amount'].mean()
print(avgg)
```

- Sort customers by total spending

```python
srt = amt.sort_values(ascending = False)
```

```python
print(srt)
```
