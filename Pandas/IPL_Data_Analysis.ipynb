# Most Impactful Batter in IPL

```python
bat = ipl.groupby('batter').agg(
    total_runs = ('runs_batter' , 'sum'),
    ball_faced = ('valid_ball' , 'sum'),
    matches_played = ('match_id' , 'nunique'),
    got_out = ('striker_out' , lambda x : ((x == True)).sum())
).reset_index()

team = ipl.groupby( 'batting_team').agg(
    total_runs_byTeam = ('runs_total' , 'sum'),
    matches_played_byTeam = ('match_id' , 'nunique')
).reset_index()

bat = bat[((bat['ball_faced'] > 300) & (bat['matches_played'] > 30) & (bat['total_runs'] > 500))]

bat['strike_rate'] = (bat['total_runs'] / bat['ball_faced'])*100
bat['average'] = bat['total_runs'] / bat['got_out']
bat['runs_per_ball'] = bat['total_runs'] / bat['ball_faced']
bat['runs_per_match'] = bat['total_runs'] / bat['matches_played']

bat['sr_norm'] = bat['strike_rate'] / bat['strike_rate'].max()
bat['avg_norm'] = bat['average'] / bat['average'].max()
bat['rpm_norm'] = bat['runs_per_match'] / bat['runs_per_match'].max()
# impact = (0.4 * sr_norm) + (0.3 * avg_norm) + (0.3 * rpm_norm)
bat['impact'] = (0.4 * bat['sr_norm']) + (0.3 * bat['avg_norm']) + (0.3 * bat['rpm_norm'])

batter_match_runs = ipl.groupby(['batter', 'match_id'])['runs_batter'].sum().reset_index()
consistency = batter_match_runs.groupby('batter')['runs_batter'].agg(['mean','var']).reset_index()
bat['consistency'] = consistency['var'] / consistency['var'].max()
consistency['std'] = consistency['var'] ** 0.5
bat['cv'] = consistency['std'] / consistency['mean']

bat.sort_values(by=['impact','cv','strike_rate','total_runs'] , ascending=[False,True,False,False]).head(30)
# bat.sort_values(by='cv' , ascending=True)
# comb = bat.merge(team , left_on=['match_id'] , right_on=['match_id'])
# comb

# comb['contri'] = (comb['total_runs'] / comb['total_runs_byTeam'])*100

# comb.sort_values(by='contri' , ascending=False).head(30)
```

```python
batter_match_runs = ipl.groupby(['batter', 'match_id', 'batting_team'])['runs_batter'].sum().reset_index()
team_match_runs = ipl.groupby(['match_id', 'batting_team'])['runs_total'].sum().reset_index()
batter_match_runs = batter_match_runs.merge(team_match_runs , on=['match_id', 'batting_team'])

batter_match_runs['contribution'] = (batter_match_runs['runs_batter'] / batter_match_runs['runs_total']) * 100

final = batter_match_runs.groupby('batter').agg(
    avg_contribution = ('contribution', 'mean'),
    max_contribution = ('contribution', 'max'),
    matches = ('match_id', 'nunique')
).reset_index()

final.sort_values(by='max_contribution' , ascending=False).head(30)
```

```python
plt.figure(figsize=(35,7))
plt.scatter(bat['average'] , bat['strike_rate'] , c='r')
plt.title("Most Impactful Batsman")
plt.xlabel("average" , fontsize=20)
plt.ylabel("Strik Rate",fontsize=20)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
```

```python
def category(row):
    if row['strike_rate'] >= 150 and row['average'] >= 35:
        return 'Elite'
    elif row['strike_rate'] >= 150 and row['average'] < 35:
        return 'Aggressive'
    elif row['strike_rate'] < 150 and row['average'] >= 35:
        return 'Anchor'
    elif 130 <= row['strike_rate'] < 150 and 25 <= row['average'] < 35:
        return 'Average'
    else:
        return 'Weak'

bat['category'] = bat.apply(category, axis=1)


plt.figure(figsize=(12,7))

for cat, data in bat.groupby('category'):
    plt.scatter(data['average'], data['strike_rate'], label=cat)

top = bat.sort_values(by='impact', ascending=False).head(30)

for i in range(len(top)):
    plt.text(
        top['average'].iloc[i],
        top['strike_rate'].iloc[i],
        top['batter'].iloc[i],
        fontsize=7
    )

plt.axvline(x=35, linestyle='--')
plt.axhline(y=150, linestyle='--')
plt.axhline(y=130, linestyle='--') 
plt.axvline(x=25, linestyle='--')

plt.legend()
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.title("Most Impactful Batsmen")

plt.show()
```
