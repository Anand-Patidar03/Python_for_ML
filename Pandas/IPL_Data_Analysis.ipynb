## Highest Strike Rate of Batsman

```python
ttl_runs = ipl.groupby('batter')['runs_batter'].sum()
ttl_ball_faced = ipl.groupby('batter')['balls_faced'].sum()

mask = ttl_runs > 300

runs_filtered = ttl_runs[mask]
balls_filtered = ttl_ball_faced[mask]

sr = (runs_filtered / balls_filtered) * 100

highest_SR = sr.sort_values(ascending=False)
```

```python
highest_SR
```

## Best economy bowlers (min 200 balls)

```python
df_valid = ipl[ipl['valid_ball'] == 1]
ttl_ball_bowled = df_valid.groupby('bowler')['ball'].count().sort_values(ascending=False)
ttl_runs_conced = ipl.groupby('bowler')['runs_bowler'].sum().sort_values(ascending=False)

ttl_overs= (ttl_ball_bowled/6)

mask = ttl_ball_bowled>300
ttl_runs_conced_final = ttl_runs_conced[mask]
ttl_overs_final = ttl_overs[mask]

economy = ttl_runs_conced_final/ttl_overs_final

best_eco = economy.sort_values(ascending = True)
```

```python
best_eco
```

## Most sixes and fours by players

## Orange Cap winner per season

```python
all_batter_inaSeason = ipl.groupby(['season' , 'batter'])['runs_batter'].sum().reset_index(name = 'total_runs_inASeason')

all_batter_inaSeason

idx = all_batter_inaSeason.groupby('season')['total_runs_inASeason'].idxmax()

org_cap = all_batter_inaSeason.loc[idx]
```

```python
org_cap
```

# Purple Cap winner per season

```python
all_bowler_inaSeason = ipl.groupby(['season','bowler'])['bowler_wicket'].sum().reset_index(name = 'total_wickets_inASeason')

all_bowler_inaSeason

idx = all_bowler_inaSeason.groupby('season')['total_wickets_inASeason'].idxmax()

purple_cap = all_bowler_inaSeason.loc[idx]
purple_cap
```
